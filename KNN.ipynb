## K-Means

Importo las dependencias

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets._samples_generator import make_blobs
```

Genero una distribuci√≥n de 4 clusters, sin etiquetas.

```python
X, y_true = make_blobs(n_samples=400, centers=4,
                       cluster_std=0.60, random_state=0)
X = X[:, ::-1] # flip axes for better plotting
plt.scatter(X[:,0], X[:,1])
```

Si a este dataset yo le aplico K-Means me va a formar clusters de acorde al valor K, es decir yo le digo al algor√≠tmo de este distribuci√≥n busca la mejor agrupaci√≥n de 4 elementos. K-Means es aprendizaje NO supervizado ya que no se le asinga ninguna etiqueta a los datos. Justamente la idea del algor√≠tmo es poder realizar esas etiquetas.

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

El punto en el centro de cada cluster representa el valor medio convergente de los 4 clusters que le indicamos. Con ese valor medio se calcula un radio en el que se incluye a los elemntos del cluster.
